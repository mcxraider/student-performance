## 3. **Data Cleaning**

### a. **Handling Missing Values**
- **Identify Missing Data:** `df.isnull().sum()`, missing value patterns.
- **Assess Missingness Mechanism:** MCAR, MAR, MNAR.
- **Strategies:**
  - **Deletion:** Remove rows or columns with excessive missingness.
  - **Imputation:**
    - **Numeric:** Mean, median, mode, interpolation.
    - **Categorical:** Mode, a new category.
    - **Advanced:** KNN imputation, MICE.

### b. **Handling Duplicates**
- **Identify Duplicates:** `df.duplicated()`.
- **Decide on Action:** Remove duplicates (`df.drop_duplicates()`), keep based on criteria.

### c. **Correcting Data Types**
- **Ensure Proper Types:** Convert to appropriate types (e.g., integers, floats, dates).
- **Use Pandas Methods:** `pd.to_datetime()`, `astype()`.

### d. **Handling Outliers**
- **Identify Outliers:**
  - **Statistical Methods:** Z-score, IQR.
  - **Visualization:** Box plots, scatter plots.
- **Decide on Action:**
  - **Removal:** If they are errors.
  - **Transformation:** Log, square root.
  - **Capping:** Winsorization.

### e. **Dealing with Inconsistent Data**
- **Standardize Formats:**
  - **Dates:** Ensure consistent date formats.
  - **Strings:** Consistent casing, trimming whitespace.
- **Correct Inconsistencies:** Merge similar categories, correct typos.

### f. **Handling Categorical Variables**
- **Standardize Categories:** Uniform naming conventions.
- **Consolidate Rare Categories:** Combine infrequent categories into 'Other'.
- **Check for Cardinality:** High-cardinality categories may require special handling.

### g. **Addressing Data Entry Errors**
- **Identify Anomalies:** Implausible values, impossible dates.
- **Correct or Remove:** Based on context and feasibility.

```python
# Count plot of each level of income, and at each level of income what is the demographic of the parents
low_income  = df[df["Family_Income"] == 'Low']
med_income  = df[df["Family_Income"] == 'Medium']
high_income = df[df["Family_Income"] == 'High']

# Define the specific order for the categorical variable
category_order = ['High School', 'College', 'Postgraduate']

# Ensure that the 'Parental_Education_Level' column in all datasets follows this order
low_income['Parental_Education_Level'] = pd.Categorical(low_income['Parental_Education_Level'], categories=category_order, ordered=True)
med_income['Parental_Education_Level'] = pd.Categorical(med_income['Parental_Education_Level'], categories=category_order, ordered=True)
high_income['Parental_Education_Level'] = pd.Categorical(high_income['Parental_Education_Level'], categories=category_order, ordered=True)

# Create a figure with 3 subplots (1 row, 3 columns)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot the low-income data on the first subplot
sns.countplot(ax=axs[0], data=low_income, x="Parental_Education_Level", order=category_order)
axs[0].set_title("Low Income Family Distribution")

# Plot the medium-income data on the second subplot
sns.countplot(ax=axs[1], data=med_income, x="Parental_Education_Level", order=category_order)
axs[1].set_title("Medium Income Family Distribution")

# Plot the high-income data on the third subplot
sns.countplot(ax=axs[2], data=high_income, x="Parental_Education_Level", order=category_order)
axs[2].set_title("High Income Family Distribution")

# Rotate x-axis labels for readability
for ax in axs:
    ax.tick_params(axis='x', rotation=45)

# Show the plots
plt.tight_layout()
plt.show()
```

So from these count plots, its fair to say that the low and medium income families mainly come from high school background.
